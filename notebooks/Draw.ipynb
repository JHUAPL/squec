```python
import matplotlib.pyplot as plt
from squec.surface import VectorIJ, Surface
```

# Common Surface Codes

## Surface-17

```python
s = Surface(VectorIJ(7, 7))
plt.figure(figsize=(7, 7))
s.draw()
```

## Surface-49

```python
s = Surface(VectorIJ(11, 11))
plt.figure(figsize=(9, 9))
s.draw()
```

## Surface-241 (1111)

```python
s = Surface(VectorIJ(23, 23))
plt.figure(figsize=(18, 18))
s.draw()
```

# Weird Surface Codes

## Offset Origin

```python
s = Surface(VectorIJ(5, 5), origin_ij=(1, 1))
plt.figure(figsize=(4, 4))
s.draw()
```

## Long and Narrow

```python
s = Surface(VectorIJ(27, 11))
plt.figure(figsize=(27, 11))
s.draw()
```
