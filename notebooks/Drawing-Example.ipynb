```python
import matplotlib.pyplot as plt
from squec.surface import VectorIJ, Surface
from squec import squec_solve
```

```python
# s = Surface(VectorIJ(7, 11))
s = Surface(VectorIJ(7, 7)) # Surface-17
s.flip_ancilla(VectorIJ(4, 4))
s.flip_ancilla(VectorIJ(2, 2))
s.flip_ancilla(VectorIJ(0, 4))
s.flip_ancilla(VectorIJ(6, 2))

solve = squec_solve(s)

# Side-by-side pre and post-solve:
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
s.draw(draw_kwargs=dict(ax=ax[0]))
ax[0].set_title("Initial State", fontsize=16, family="Arial")
ax[0].axis("off")
s.highlight_vertices(solve)
s.draw(draw_kwargs=dict(ax=ax[1]))
ax[1].set_title("Solve", fontsize=16, family="Arial")
ax[1].axis("off")
s.draw()
```
